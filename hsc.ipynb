```python
# Algorithm: Matrix-Multiply(A, B)
# Input: Matrix A of size m x n, Matrix B of size  n x p
# Output: Matrix C of size m x p

# 1. if columns(A) ≠ rows(B) then
# 2.     return "Matrices cannot be multiplied"
# 3. 
# 4. Initialize C[m][p] with all elements = 0
# 5. 
# 6. for i = 1 to m do
# 7.     for j = 1 to p do
# 8.         for k = 1 to n do
# 9.             C[i][j] = C[i][j] + A[i][k] * B[k][j]
# 10.        end for
# 11.    end for
# 12. end for
# 13. 
# 14. return C
```

```python


import numpy as np

A = np.random.uniform(low=0, high=1, size=(300, 400))
B = np.random.uniform(low=0, high=1, size=(400, 500))
```

```python
import time

def timeit(func):
    """Decorator that times the execution of a function"""
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()
        result = func(*args, **kwargs)
        end_time = time.perf_counter()
        elapsed_time = 1_000*(end_time - start_time)
        print(f"{func.__name__} took {elapsed_time:.6f} miliseconds")
        return result
    return wrapper

    
```

```python
import torch

A_cuda = torch.from_numpy(A).cuda()
B_cuda = torch.from_numpy(B).cuda()


@timeit
def cuda_matmul(A,B):
    return torch.mm(A_cuda, B_cuda)  # or A @ B

C_cuda = cuda_matmul(A,B)
```

```python


@timeit
def matrix_multiplication(A,B):

    if A.shape[1] == B.shape[0]:

        C = np.zeros((A.shape[0],B.shape[1]))

        for i in range(A.shape[0]):
            for j in range(B.shape[1]):
                for k in range(B.shape[0]):
                    C[i][j] = C[i][j] + A[i][k] * B[k][j]

        return C
    
    else:

        raise ValueError

C_loop = matrix_multiplication(A,B)

```

```python
@timeit
def numpymatmul(A,B):
    return A@B

C_numpy0 = numpymatmul(A,B)
```

```python
@timeit
def numpymatmul2(A,B):
    return np.matmul(A,B)

C_numpy1 = numpymatmul2(A,B)
```

```python
np.allclose(C_cuda.cpu().numpy(),C_loop)
```
